## Review: Python Lists

- Zero-indexed sequence of arbitrary Python values.
- Slicing syntax: `l[start:stop:step]` copies elements at regular intervals from `start` to `stop`.
- Efficient (`O(1)`) appends and removes from end.
- Comprehension syntax: `[f(x) for x in l if cond(x)]`.

```python
%%fortran
subroutine matmul(A, B, C)
    implicit none
    real, dimension(:, :), intent(in) :: A, B
    real, dimension(size(A, 1), size(B, 2)) :: C
    integer :: i, j, k

    ! Asegurar que las dimensiones de las matrices sean correctas para la multiplicación
    if (size(A, 2) /= size(B, 1)) then
        print *, "No se pueden multiplicar las matrices. Las dimensiones no coinciden."
        return
    endif

    ! Inicializar la matriz C
    C = 0.0

    ! Realizar la multiplicación de matrices
    do i = 1, size(A, 1)
        do j = 1, size(B, 2)
            do k = 1, size(A, 2)
                C(i, j) = C(i, j) + A(i, k) * B(k, j)
            enddo
        enddo
    enddo

end subroutine matmul

program test_matmul
    implicit none
    real, dimension(:, :), allocatable :: A, B, C
    integer :: fila_A, col_A, fila_B, col_B, i, j

    ! Definir las dimensiones de las matrices
    fila_A = 600
    col_A = 100
    fila_B = 100
    col_B = 600

    ! Asignar dimensiones a las matrices
    allocate(A(fila_A, col_A))
    allocate(B(fila_B, col_B))

    ! Inicializar las matrices A y B
    A = reshape([(i, i = 1, fila_A*col_A)], [fila_A, col_A])
    B = reshape([(i, i = 1, fila_B*col_B)], [fila_B, col_B])

    ! Llamar a la subrutina para multiplicar las matrices
    call matmul(A, B, C)

    ! Imprimir las matrices originales y el resultado de la multiplicación
    print *, "Matriz A:"
    do i = 1, fila_A
        print *, (A(i, j), j = 1, col_A)
    end do

    print *, "Matriz B:"
    do i = 1, fila_B
        print *, (B(i, j), j = 1, col_B)
    end do

    print *, "Resultado de la multiplicación (Matriz C):"
    do i = 1, fila_A
        print *, (C(i, j), j = 1, col_B)
    end do

    ! Liberar la memoria asignada a las matrices
    deallocate(A)
    deallocate(B)

end program test_matmul
```

**My own example 15 - Running 10 times the  Fortran subroutine matmul(A,B) with 600x100 and 100x600 matrices and adding Fortran magic average execution time to the data frame**

```python
import numpy as np
import pandas as pd
import time
from ctypes import CDLL, POINTER, c_int, c_double

# Load the compiled shared library
matmul_lib = CDLL("./matmul.so")

# Define the function signature for the Fortran subroutine
matmul_lib.matmul_.argtypes = [
    POINTER(c_double), POINTER(c_double), POINTER(c_double),
    c_int, c_int, c_int
]

# Define a function to compute the average execution time
def average_execution_time(A, B, m, n, p, num_iterations=10):
    execution_times = []
    for _ in range(num_iterations):
        start_time = time.time()
        matmul_lib.matmul_(A.ctypes.data_as(POINTER(c_double)),
                           B.ctypes.data_as(POINTER(c_double)),
                           np.zeros((m, p), dtype=np.float64).ctypes.data_as(POINTER(c_double)),
                           m, n, p)
        end_time = time.time()
        execution_time = end_time - start_time
        execution_times.append(execution_time)
    return np.mean(execution_times)

# Create a DataFrame to store the results
data = {'Matrix_Size': [], 'Execution_Time': []}

# Generate and multiply matrices 10 times
for _ in range(10):
    m, n, p = 600, 100, 600
    A = np.random.rand(m, n)
    B = np.random.rand(n, p)
    execution_time = average_execution_time(A, B, m, n, p)
    data['Matrix_Size'].append((m, n, p))
    data['Execution_Time'].append(execution_time)

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print(df)
```

```python
%%fortran
subroutine matmul(A, B, C)
    implicit none
    real, dimension(:, :), intent(in) :: A, B
    real, dimension(size(A, 1), size(B, 2)) :: C
    integer :: i, j, k

    ! Asegurar que las dimensiones de las matrices sean correctas para la multiplicación
    if (size(A, 2) /= size(B, 1)) then
        print *, "No se pueden multiplicar las matrices. Las dimensiones no coinciden."
        return
    endif

    ! Inicializar la matriz C
    C = 0.0

    ! Realizar la multiplicación de matrices
    do i = 1, size(A, 1)
        do j = 1, size(B, 2)
            do k = 1, size(A, 2)
                C(i, j) = C(i, j) + A(i, k) * B(k, j)
            enddo
        enddo
    enddo

end subroutine matmul

program test_matmul
    implicit none
    real, dimension(:, :), allocatable :: A, B, C
    integer :: fila_A, col_A, fila_B, col_B, i, j, n

    ! Definir las dimensiones de las matrices
    fila_A = 600
    col_A = 100
    fila_B = 100
    col_B = 600
    n = 10 ! Número de veces para multiplicar las matrices

    ! Asignar dimensiones a las matrices
    allocate(A(fila_A, col_A))
    allocate(B(fila_B, col_B))

    ! Inicializar las matrices A y B
    A = reshape([(i, i = 1, fila_A*col_A)], [fila_A, col_A])
    B = reshape([(i, i = 1, fila_B*col_B)], [fila_B, col_B])

    do i = 1, n
        ! Llamar a la subrutina para multiplicar las matrices
        call matmul(A, B, C)

        ! Imprimir las matrices originales y el resultado de la multiplicación
        print *, "Iteración", i
        print *, "Resultado de la multiplicación (Matriz C):"
        do j = 1, fila_A
            print *, (C(j, k), k = 1, col_B)
        end do
    end do

    ! Liberar la memoria asignada a las matrices
    deallocate(A)
    deallocate(B)

end program test_matmul
```

## Why is the Python Version so Much Slower?

```python
# Dynamic typing.
def mul_elemwise(xs, ys):
    return [x * y for x, y in zip(xs, ys)]

mul_elemwise([1, 2, 3, 4], [1, 2 + 0j, 3.0, 'four'])
#[type(x) for x in _]
```

```python
# Interpretation overhead.
source_code = 'a + b * c'
bytecode = compile(source_code, '', 'eval')
import dis; dis.dis(bytecode)
```

## Why is the Python Version so Slow?
- Dynamic typing means that every single operation requires dispatching on the input type.
- Having an interpreter means that every instruction is fetched and dispatched at runtime.
- Other overheads:
  - Arbitrary-size integers.
  - Reference-counted garbage collection.

> This is the paradox that we have to work with when we're doing scientific or numerically-intensive Python. What makes Python fast for development -- this high-level, interpreted, and dynamically-typed aspect of the language -- is exactly what makes it slow for code execution.

- Jake VanderPlas, [*Losing Your Loops: Fast Numerical Computing with NumPy*](https://www.youtube.com/watch?v=EEUXKG97YRw)

```python
# What's the regional average price of a HASS avocado every day?
hass = avocados[avocados.Variety == 'HASS']
hass.groupby(['Date', 'Region'])['Weighted Avg Price'].mean().unstack().ffill().plot();
```

```python
def _organic_spread(group):

    if len(group.columns) != 2:
        return pd.Series(index=group.index, data=0.0)

    is_organic = group.columns.get_level_values('Organic').values.astype(bool)
    organics = group.loc[:, is_organic].squeeze()
    non_organics = group.loc[:, ~is_organic].squeeze()
    diff = organics - non_organics
    return diff

def organic_spread_by_region(df):
    """What's the difference between the price of an organic
    and non-organic avocado within each region?
    """
    return (
        df
        .set_index(['Date', 'Region', 'Organic'])
         ['Weighted Avg Price']
        .unstack(level=['Region', 'Organic'])
        .ffill()
        .groupby(level='Region', axis=1)
        .apply(_organic_spread)
    )
```
